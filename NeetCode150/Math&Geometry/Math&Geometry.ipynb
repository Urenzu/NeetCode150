# Rotate image

```python
class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        l, r = 0, len(matrix) - 1

        while l < r:
            for i in range(r - l):
                top , bottom = l, r

                # Save the topleft value
                topLeft = matrix[top][l + i]

                # Move bottom left into top left
                matrix[top][l + i] = matrix[bottom - i][l]

                # Move bottom right into bottom left
                matrix[bottom - i][l] = matrix[bottom][r - i]

                # Move top right into bottom right
                matrix[bottom][r - i] = matrix[top + i][r]

                # Move top left into top right
                matrix[top + i][r] = topLeft
                
                # We use i to increment the layers in order to rotate all numbers on the outer layer
            
            # This is used to switch our range to the inner layer
            r -= 1
            l += 1
```

# Spiral Matrix

```python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        res = []
        left, right = 0, len(matrix[0])
        top, bottom = 0, len(matrix)

        while left < right and top < bottom:
            # get every i in the top row
            for i in range(left, right):
                res.append(matrix[top][i])
            top += 1
            # get every i in the right col
            for i in range(top, bottom):
                res.append(matrix[i][right - 1])
            right -= 1
            if not (left < right and top < bottom):
                break
            # get every i in the bottom row
            for i in range(right - 1, left - 1, -1):
                res.append(matrix[bottom - 1][i])
            bottom -= 1
            # get every i in the left col
            for i in range(bottom - 1, top - 1, -1):
                res.append(matrix[i][left])
            left += 1

        return res
```

# Set Matrix Zeroes

```python
class Solution:
    def setZeroes(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        rows, cols = len(matrix), len(matrix[0]) # get the number of rows and columns in the matrix
        zero = False # initialize a flag variable to keep track of the first row
        
        # Determine which rows/cols need to be zero
        for r in range(rows):
            for c in range(cols):
                if matrix[r][c] == 0: # if the element at (r, c) is 0
                    matrix[0][c] = 0 # mark the first element of the column as 0

                    if r > 0:
                        matrix[r][0] = 0 # mark the first element of the row as 0 (if not the first row)
                    else:
                        zero = True # if the first row has a 0, mark the flag variable as True
                    
        # Set the elements to 0 using the first row and column
        for r in range(1, rows):
            for c in range(1, cols):
                if matrix[0][c] == 0 or matrix[r][0] == 0: # if either the first element of the column or row is 0
                    matrix[r][c] = 0 # set the current element to 0
        
        # Handle the first row and column separately
        if matrix[0][0] == 0: # if the first element of the matrix is 0
            for r in range(rows):
                matrix[r][0] = 0 # set the elements in the first column to 0
        
        if zero: # if the flag variable is True (i.e., if the first row had a 0)
            for c in range(cols):
                matrix[0][c] = 0 # set the elements in the first row to 0
```

# Happy Number

```python
class Solution:
    def isHappy(self, n: int) -> bool:
        visit = set() # visit hashset

        while n not in visit:
            visit.add(n)
            n = self.squaresum(n) # Keep iterating through all sums of squares and making a list within our set

            if n == 1: # once we find a happy number (always repeating)
                return True

        return False

    def squaresum(self, n: int) -> int:
        output = 0

        while n:
            digit = n % 10 # This gets the last value within the number
            digit = digit ** 2 # We square the number and add it into our output
            output += digit
            n = n // 10 # By doing // 10 we're effectively getting rid of the last value within our number... we do this so digit can access the next value, it effectively iterates through our entire value.

        return output
```

# Plus One

```python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        digits = digits[::-1]
        one, i = 1, 0
        while one: 
            if i < len(digits): # While we iterate through the array backwards
                if digits[i] == 9: # If we discover a 9
                    digits[i] = 0 # Set that index to 0
                else: # Increment by one since that's our default case. We have to increment the last digit by one. 
                    digits[i] += 1 # While we have a carry the next value will recieve the one.
                    one = 0 # At the end of our increment we set one to 0 to reset the loop.
            else: # However if our array is empty we add a digit to the array.
                digits.append(1)
                one = 0 # Terminating the loop

            i += 1 # To continue iterating through loop when needed
        
        return digits[::-1] # We reversed our array earlier so now we have to reverse it again for the correct returning format

# A better more simple solution
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        for i in range(len(digits) - 1, -1, -1):
            if digits[i] == 9:
                digits[i] = 0
            else:
                digits[i] += 1
                return digits
        digits.insert(0, 1)
        return digits
```

# Pow(x, n)

```python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        def helper(x, n):
            # Base cases:
            # If x is 0, the result is always 0
            # If n is 0, the result is always 1
            if x == 0: return 0
            if n == 0: return 1

            # Divide the power in half and recursively compute the result
            res = helper(x, n // 2)

            # Multiply the result by itself to get the result of x raised to the power of n/2
            res = res * res

            # If n is odd, multiply the result by x one more time
            if n % 2:
                return x * res
            else:
                # If n is even, just return the result
                return res 

        # Call the helper function with the absolute value of n
        res = helper(x, abs(n))

        # If n is negative, return 1 divided by the result
        if n < 0:
            return 1 / res
        else:
            # If n is non-negative, return the result as is
            return res
```

# Multiply Strings

```python
class Solution:
    def multiply(self, num1: str, num2: str) -> str:
        # If either number is 0, return 0
        if "0" in [num1, num2]:
            return "0"

        # Initialize a result list of zeros with length equal to the sum of the lengths of the input numbers
        res = [0] * (len(num1) + len(num2))

        # Reverse the input numbers to make it easier to iterate over them from right to left
        num1, num2 = num1[::-1], num2[::-1]

        # Iterate over each digit in num1 and num2 and compute the product and add it to the result list
        for i1 in range(len(num1)):
            for i2 in range(len(num2)):
                digit = int(num1[i1]) * int(num2[i2])
                res[i1 + i2] += digit
                res[i1 + i2 + 1] += res[i1 + i2] // 10  # Carry over the excess value to the next digit
                res[i1 + i2] = res[i1 + i2] % 10  # Take the modulo 10 to get the correct digit

        # Reverse the result list and remove any leading zeros
        res, beg = res[::-1], 0
        while beg < len(res) and res[beg] == 0:
            beg += 1
        res = map(str, res[beg:])

        # Concatenate the digits in the result list to get the final result as a string
        return "".join(res)
```

# Detect Squares

```python
class DetectSquares:
    def __init__(self):
        # Initialize a dictionary to store the count of each point
        self.ptsCount = defaultdict(int)
        # Initialize a list to store all the points
        self.pts = []

    def add(self, point: List[int]) -> None:
        # Increment the count of the input point in the dictionary
        self.ptsCount[tuple(point)] += 1
        # Add the input point to the list of all points
        self.pts.append(point)

    def count(self, point: List[int]) -> int:
        # Initialize a result variable to store the count of squares containing the input point
        res = 0
        # Get the x and y coordinates of the input point
        px, py = point
        # Iterate over all the points in the list
        for x, y in self.pts:
            # If the x and y coordinates of the current point are the same as the input point, skip it
            if (abs(py - y) != abs(px - x)) or x == px or y == py:
                continue
            # Compute the count of squares containing the input point and the current point, and add it to the result
            res += self.ptsCount[(x, py)] * self.ptsCount[(px, y)]
        # Return the final count of squares
        return res
```
