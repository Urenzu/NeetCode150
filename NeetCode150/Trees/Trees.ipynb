# Invert Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root: # If we don't have a root node return nothing.
            return None
        
        # Swap children
        tmp = root.left
        root.left = root.right
        root.right = tmp

        # Calling invertTree function to swap the nodes.
        self.invertTree(root.left) 
        self.invertTree(root.right)
        return root

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
        if not root:
            return None
        
        tmp = root.left
        root.left = self.invertTree(root.right)
        root.right = self.invertTree(tmp)

        return root
```

# Maximum Depth of Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

# Recursive approach, goes from first node and checks the depth on left side, then checks depth on right side. DFS
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0

        return 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))

class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        return 0 if not root else 1 + max(self.maxDepth(root.left), self.maxDepth(root.right))

# BFS Checks each level one at a time
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if not root:
            return 0

        level = 0
        q = deque([root]) # Set up queue with root node
        while q: 
            for i in range(len(q)): # While we go through queue pop current nodes and then search to the left and right for more nodes on another level, append to queue.
                node = q.popleft()

                if node.left:
                    q.append(node.left)
                if node.right:
                    q.append(node.right)

            level += 1 # Increment level to show we hod nodes and we popped them
        return level

# Iterative preorder DFS with stack data structure
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        stack = [[root, 1]]
        res = 0

        while stack:
            node, depth = stack.pop()

            if node: # While we have nodes to the left or right append them to the stack and increase our depth level, everytime we increase our depth level res gets a higher value.
                res = max(res, depth)
                stack.append([node.left, depth + 1])
                stack.append([node.right, depth + 1])
        
        return res
```

# Diameter of Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
        res = [0]

        def dfs(root):
            if not root: # If we have no nodes return -1
                return -1
            left = dfs(root.left) # else left do a depth first search on the left and right nodes
            right = dfs(root.right)

            res[0] = max(res[0], 2 + left + right) # set result = the max of either res or 2 + left + right

            return 1 + max(left, right) # Return the max but add one for the root node
        
        dfs(root) # Call dfs function and then return the diameter
        return res[0]
```

# Balanced Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isBalanced(self, root: Optional[TreeNode]) -> bool:

        def dfs(root):
            if not root: return [True, 0]

            left, right = dfs(root.left), dfs(root.right) # Depth first search to left and right nodes from the root node.
            balance = (left[0] and right[0] and abs(left[1] - right[1])) <= 1 # From the very bottom going upwards we from the root node take the child nodes and subtract their values. While the height difference is less than or equal to 1 then they're balanced. We need to compare all parent nodes and root nodes.
            return [balanced, 1 + max(left[1], right[1])] # Returning max height plus root node and whether it's balanced.

        return dfs(root)[0] # Returning pair of values from dfs, but we're only interested in the first value which indicates true or false for if the tree is balanced.

```

# Same Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
        if not p and not q: # If both trees don't have any nodes they're equal
            return True
        if not p or not q or p.val != q.val: # If one or the other has a node but the other doesn't, or the root nodes are not equal.
            return False

        return (self.isSameTree(p.left, q.left) and self.isSameTree(p.right, q.right)) # checking left and right nodes one each tree  and seeing if values and positions are the same.
```

# Lowest Common Ancestor of a Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        if p.val > root.val and q.val > root.val: # While p and q are greater than root value search the lowest common ancestor to the right
            return self.lowestCommonAncestor(root.right, p, q)
        elif p.val < root.val and q.val < root.val: # While p and q are less than root value search the lowest common ancestor to the left
            return self.lowestCommonAncestor(root.left, p, q)
        else:
            return root
        
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        
        # Eventually while traversing the tree we will reach a null leaf node which will make our while loop false prompting us to return the lowest common ancestor root node.
        while root:
            if p.val > root.val and q.val > root.val:
                root = root.right
            elif p.val < root.val and q.val < root.val:
                root = root.left
            else:
                return root
```

# Binary Tree Level Order Traversal

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        # Use BFS level order traversal of a binary tree implemented using a queue in linear time

        res = []

        q = collections.deque() # How to initialize a queue
        q.append(root)

        while q: # Run our queue while it isn't empty
            q_length = len(q) # iterating through one level
            level = []
            for i in range(q_length):
                node = q.popleft()
                if node: # If the node isn't null so if we have a parent node append our left and right children nodes if they exist.
                    level.append(node.val)
                    q.append(node.left)
                    q.append(node.right)
            if level: # Making sure the level isn't empty. We don't want empty levels full of null nodes.
                res.append(level) # Everytime we go through a section of nodes append it to our result list.

        return res


```

# Binary Tree Right Side View

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right

class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
        def dfs(node, depth, res):
            if not node: # If we don't have a node.
                return
            
            if depth == len(res):
                res.append(node.val)
            
            dfs(node.right, depth + 1, res)
            dfs(node.left, depth + 1, res)
        
        res = []
        dfs(root, 0, res)
        return res

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def rightSideView(self, root: Optional[TreeNode]) -> List[int]:

        res = []
        q = collections.deque([root]) # Our queue could have null values, we'll handle that later.

        while q:
            right = None
            q_length = len(q)

            for i in range(q_length): # Going through the length of our levels of the tree starting from 0, our first node.
                node = q.popleft() # For every tree remove the parent node from the queue
                if node: # If we had a parent node than append the left and right child nodes
                    right = node 
                    q.append(node.left) # Since left node gets appended first it is the first to be removed, the last node to be removed (the rightmost node) is the only node which will be added to our result array
                    q.append(node.right)
            if right:
                res.append(right.val)
            
        return res
```

# Count Good Nodes in Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def goodNodes(self, root: TreeNode) -> int:
        # Pre-order traversal, on dfs process each node before we move on to another subtree

        def dfs(node, maxVal):
            if not node:
                return 0
            
            res = 1 if node.val >= maxVal else 0 # As we traverse the tree check to see if the node value is bigger than our max node value so far. If so then it's a good node else not a good node.
            maxVal = max(maxVal, node.val)
            res += dfs(node.left, maxVal) # Incrementing our result every time we find a good left or right node.
            res += dfs(node.right, maxVal)
            return res

        return dfs(root, root.val) # We call using the root node and pin it to the maximum value seen so far
```

# Validate Binary Search Tree

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def isValidBST(self, root: Optional[TreeNode]) -> bool:

        def valid(node, left, right):
            if not node: # Because technically an empty binary search tree is still a binary search tree
                return True
            
            if not(node.val < right and node.val > left): # Comparing to see if left value is less and right value is greater, if not then not a bst.
                return False
            
            return (valid(node.left, left, node.val) and valid(node.right, node.val, right)) # Call to function to validate all values on left and right. It's really easy to do this when you setup (node, left, right) when you call make sure you use -inf for left boundary and inf for right.

        return valid(root, float("-inf"), float("inf"))
```

# Kth Smallest Element in a BST

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root: TreeNode, k: int) -> int:
        stack = []
        curr = root

        while stack or curr: # While current isn't null and while stack isn't empty
            while curr:
                stack.append(curr) # Append cur node then move pointer to the left
                curr = curr.left
            curr = stack.pop()  # Pop most recent value from stack
            k -= 1
            if k == 0:
                return curr.val  # Return the value at the kth node
            curr = curr.right
```

# Construct Binary Tree From Preorder And Inorder Traversal

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
        # Pre-order traversal goes root node to left to right all the way down.
        # In-order traversal goes left tree to root node to right tree.
        if not preorder or not inorder:
            return None
        
        root = TreeNode(preorder[0]) # First value of preorder array
        mid = inorder.index(preorder[0])
        root.left = self.buildTree(preorder[1: mid + 1], inorder[:mid]) # Creating sublists to create left subtree
        root.right = self.buildTree(preorder[mid + 1:], inorder[mid + 1:]) # Creating sublists to create right subtree
        return root
```

# Binary Tree Maximum Path Sum

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        res = [root.val]

        def dfs(root):
            if not root: # If no root
                return 0
            
            left = dfs(root.left)
            right = dfs(root.right)
            left = max(left, 0)
            right = max(right, 0)

            # Compute max path sum WITH split
            res[0] = max(res[0], root.val + left + right)

            return root.val + max(left, right) # Return either the left or right max, we can't choose both because we do then we're splitting.
        
        dfs(root)
        return res[0]
```

# Serialize And Deserialize Binary Tree

```python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:
    # DFS pre-order, O(n) for both
    def serialize(self, root):
        """Encodes a tree to a single string.
        :type root: TreeNode
        :rtype: str
        """
        res = []

        def dfs(node):
            if not node:
                res.append("N") # If we have a null node return "N"
                return
            res.append(str(node.val)) # If we have a node convert the nodes value into a string dfs on all left and right nodes
            dfs(node.left)
            dfs(node.right)
        dfs(root)
        return ",".join(res) # Delimiter for seperating values

    def deserialize(self, data):
        """Decodes your encoded data to tree.
        :type data: str
        :rtype: TreeNode
        """
        vals = data.split(",") # Split on the comma, this gets rid of all the commas and just maintains a list full of numbers
        self.i = 0 # Starting at the first value of our list of values

        def dfs():
            if vals[self.i] == "N": # If our value is null return None
                self.i += 1 # We want to continue iterating through our list values
                return None
            node = TreeNode(int(vals[self.i])) # Convert string back into integer
            self.i += 1
            node.left = dfs()
            node.right = dfs()
            return node
        return dfs()
        

# Your Codec object will be instantiated and called as such:
# ser = Codec()
# deser = Codec()
# ans = deser.deserialize(ser.serialize(root)) 
```
