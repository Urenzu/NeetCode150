# Valid Palindrome

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        string = ""

        for c in s:
            if c.isalnum(): # Built in function to check if character is alphanumeric
                string += c.lower() # If alphanumeric add it to our cleaned string and convert the character to lower case.
        return string == string[::-1] # [::-1] = syntax for reversing a string

class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = "".join(c.lower() for c in s if c.isalnum())
        return s == s[::-1]

class Solution:
    def isPalindrome(self, s: str) -> bool:
        return ''.join(c.lower() for c in s if c.isalnum()) == ''.join(c.lower() for c in s if c.isalnum())[::-1]

class Solution:
    def isPalindrome(self, s: str) -> bool:
        left, right = 0, len(s) - 1

        while left < right: # Progress loop until the hit eachother
            # It is important to have the left < right, and right > left checks, so we don't decrement or increment characters past our other pointer. Stay in proper range.
            # Make sure to use self. to call a function inside of an object.
            while left < right and not self.alpha(s[left]): # While left value is less than right value and  the character is not alphanumeric increment to next character.
                left += 1
            while right > left and not self.alpha(s[right]): # While right value is greater than left value and  the character is not alphanumeric decrement to next character.
                right -= 1
            if s[left].lower() != s[right].lower(): # Return false if the characters aren't equal
                return False
            left, right = left + 1, right - 1 # Move pointers before progressing while loop
        return True

        
        # Using ord to get the ascii value and then we're seeing if the character is within the ascii character ranges. 
        # These ranges will declare whether or not the character is alphanumeric.
    def alpha(self, c):
        return (ord('A') <= ord(c) <= ord('Z') or
                ord('a') <= ord(c) <= ord('z') or
                ord('0') <= ord(c) <= ord('9'))
```

# Two Sum II - Input Array Is Sorted

```python
class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l = 0 
        r = len(numbers) - 1

        while l != r:
            sum = numbers[l] + numbers[r]

            if sum > target: # If sum is greater than target decrement.
                r -= 1
            elif sum < target: # If sum is less than target increment.
                l += 1  
            else: # When sum == target
                return [l + 1, r + 1]

class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        l, r = 0, len(numbers) - 1

        while numbers[l] + numbers[r] != target:
            if numbers[l] + numbers[r] > target:
                r -= 1
            else:
                l += 1
        
        return [l + 1, r + 1]

class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        return [[i+1,j+1] for i in range(len(numbers)) for j in range(i+1,len(numbers)) if numbers[i] + numbers[j] == target][0]

class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        num_dict = {}
        for i, num in enumerate(numbers):
            complement = target - num
            if complement in num_dict:
                return [num_dict[complement]+1, i+1]
            num_dict[num] = i
```

# 3Sum

```python
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        res = []
        nums.sort() #Sorts the input array from smallest to greatest value.
        
        for i, a in enumerate(nums): # Means to iterate through the index and the value.
            if i > 0 and a == nums[i-1]: # This makes it so we don't use the same value twice. If the right value is the same as the left value, then skip.
                continue
            
            l, r = i + 1, len(nums) - 1 # l = i + 1, r = len(nums) - 1
            while l < r:
                threeSum = a + nums[l] + nums[r] # a + nums[1] + nums[2] = three sum
                if threeSum > 0: # If too large shift right pointer left
                    r -= 1
                elif threeSum < 0: # If too small shift left pointer right
                    l += 1
                else: # We append the series of three values for threesum but we want more combinations of values and can't have duplicate combinations so if we have duplicate left values we increment our left pointer right.
                    res.append([a, nums[l], nums[r]])
                    l += 1
                    while nums[l] == nums[l - 1] and l < r:
                        l += 1
                    # We don't do this to the right value because we're already getting a different combination by changing the left value in the series and eventually the left values will catch up to the right values.
        return res
```

# Trapping Rain Water

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        if not height: return 0 # If the height list is empty have it trap 0 water.

        l, r = 0, len(height) - 1
        leftMax, rightMax = height[l], height[r] # Assign left and right index values to two variables.
        res = 0

        while l < r:
            if leftMax < rightMax: # Increment the smaller maximum value of the two maxes as we go through the array using two pointers.
                l += 1
                leftMax = max(leftMax, height[l]) # Get the max by comparing current left height value by previous max left value.
                res += leftMax - height[l] # Add the remainder of the left max value subtracted by the current height on the left pointer.
            else:
                r -= 1 # If rightMax is greater than leftMax then do the same thing but on the right side. 
                rightMax = max(rightMax, height[r])
                res += rightMax - height[r]
                
        return res
```
